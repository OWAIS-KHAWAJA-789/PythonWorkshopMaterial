## List

Lists are very similar to strings, except that each element can be of any type.

The syntax for creating lists in Python is `[...]`:

```python
l = [1,2,3,4]
print (type(l))
print (l)
```

We can use the same slicing techniques to manipulate lists as we could use on strings:

```python
print (l)

print (l[0])

print (l[1:3]) 

print (l[::2])
```

Elements in a list do not all have to be of the same type:

```python
l = [1, 'a', 1.0, 1-1j]

print (l) 
```

Lists play a very important role in Python. For example they are used in loops and other flow control structures (discussed below). There are a number of convenient functions for generating lists of various types, for example the range function:

```python
start = 10
stop = 30
step = 2

list1 = list(range(start, stop, step))

print (list1)
```

```python
print (list(range(-10, 10)))
```

## Tuples

Tuples are like lists, except that they cannot be modified once created, that is they are *immutable*. 

In Python, tuples are created using the syntax `(..., ..., ...)`, or even `..., ...`:

```python
point = (10,)

print (point, type(point))
```

```python
point = 10, 20

print (point, type(point))
```

```python
x, y = point

print ("x =", x)
print ("y =", y)
```

```python
# tuple with mixed datatypes
my_tuple = (10, "Hello World", 10.4)
print(my_tuple)
```

```python
tupl = ("hello",)
print(type(tupl))
```

```python
# Accessing tuple elements using indexing
my_tuple = ('a','s','d','f','g')

print(my_tuple[0])   # 'a' 
print(my_tuple[4])   # 'g'
#negative indexing
print(my_tuple[-2])  #'f'
# nested tuple
n_tuple = ("hello", [8, 4, 6], (1, 2, 3))

# nested index
print(n_tuple[0][4])       # 'o'
print(n_tuple[1][1])       # 4

```

If we try to assign a new value to an element in a tuple we get an error:

```python
# my_tuple[1] = 9
#can't delete items
```

## Dictionaries

Dictionaries are also like lists, except that each element is a key-value pair. The syntax for dictionaries is `{key1 : value1, ...}`: 
 Dictionaries do not allow duplicates.

```python
params = {"parameter1" : 1.0,
          "parameter2" : 'A',
          "parameter3" : 3,}

print (type(params))
print (params)
```

```python
print (params["parameter1"])
print (params["parameter2"]) 
print (params["parameter3"])
```

```python
#Updating
params["parameter1"] = "A"
params["parameter2"] = "B"

# add a new entry
params["parameter4"] = "D"

print ("parameter1 = ",  params["parameter1"])
print ("parameter2 = ",  params["parameter2"])
print ("parameter3 = ",  params["parameter3"])
print ("parameter4 = ",  params["parameter4"])
```
